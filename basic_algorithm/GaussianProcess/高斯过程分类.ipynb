```python
import sys
sys.path.append('../../mnist/adaboost')
from my_get_data import *
```

```python
def gen_yfunc(tval):
    def yfunc(x):
        x1, x2 = x[:, 0], x[:, 1]
        return (x1 * x2 > tval).astype(np.int32)
    return yfunc
```

```python
dobj = DataCreater(gen_yfunc(1.5))
x, y = dobj.gen_rand_data(size=5000)
sc = plt.scatter(x[:, 0], x[:, 1], c=y, s=2)
plt.legend(*sc.legend_elements())
plt.show()
```

```python
from _GPC import *
gobj = GPC()
gobj.fit(x, y)
# 准确率
print(gobj.score(x, y))
```

```python
plot_full_region(dobj, gobj.predict, show_scatter=False, print_accuracy=True)
```
